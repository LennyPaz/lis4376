# Magic Commands

```python
# https://www.aboutdatablog.com/post/top-8-magic-commands-in-jupyter-notebook
# https://ipython.readthedocs.io/en/stable/interactive/magics.html

# Magic Commands:
# Special commands that help with running and analyzing data in Jupyter notebooks.
# Single % for one line of code, or double %% for several lines of code.
```

## 1. List all magic commands: %lsmagic

```python
%lsmagic
```

```python
# example line magic
%ls
```

## 2. Run file: %run

```python
# example run file
%run hello.py
```

## 3. Load file: %load

**Note:** %run runs file; %load loads file into cell

```python
# %load hello.py
# Note: File was loaded into cell, then cell code was modified!

# print('Hello, world')

# or, create function
def hello_test():
    print('Hello, test!')

hello_test()
```

## 4. Return execution time: %%time

**Note:** %%time is 'cell magic' and **must be** first line in Jupyter notebook cell.

```python
%%time
import random
for i in range(0, 1000000):
    random.random()

# display time to generate 1,000,000 random numbers
# Note: Placing these comments above code generates an error!
```

## 5. List all variables: %who

```python
# example
my_var1 = 1
my_var2 = 'hello'
my_var3 = True
```

```python
# List all variables
%who
```

```python
# List specific variable
%who bool
```

## 6. Display variable details: %pinfo

```python
# example (note: includes length of string)
%pinfo my_var2
```
